```python
class PlayerEarning:
  def __init__(self,name):
    self.name=name
    self.goal=0
  def calculateTotal(self,earn,goal=None):
    self.earn=earn
    self.goal=goal
    if self.goal==None:
      self.bonus=0
    if self.goal!=None and self.goal>30:
      self.bonus=(5/100)*self.earn + 10000
    if self.goal!=None and self.goal<30:
      self.bonus=(5/100)*self.earn
  def printDetails(self):
    print(f'Player name :{self.name}\nEarning without bonus:{self.earn}\nBonus:{self.bonus}')


print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()

print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()

print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
class Bird:
  def __init__(self,name,c=False):
      self.name=name
      self.c=c
  def fly(self):
    if self.c==False:
      self.t='Flightless Birds'
      print(f'{self.name} can not fly')
    else:
      print(f'{self.name} can fly')
  def setType(self,n):
    self.t=n
  def printDetail(self):
    print(f'name:{self.name}\nType:{self.t}')



ostrich = Bird('Ostrich')
duck = Bird("Duck", True)
owl = Bird('Owl', True)
print('###########################')
ostrich.fly()
duck.fly()
owl.fly()
duck.setType('Water Birds')
owl.setType('Birds of Prey')
print('=========================')
ostrich.printDetail()
print('=========================')
duck.printDetail()
print('=========================')
owl.printDetail()

```

```python
class Smartphone:
  def __init__(self,name=None):
    self.name=name
    self.d={}
    if self.name==None:
        print('Feature can not be added without phone name')
  def addFeature(self,n1,n2):
    if n1 not in self.d:
       self.d.update({n1:n2})
    else:
      if self.d[n1]!=n2:
         self.d[n1]=self.d[n1]+','+n2
      # print(n1)
      # print(n2)
  def setName(self,name):
    self.name=name
  def printDetail(self):
    print(f'Phone name:{self.name}')
    for i,j in self.d.items():
      print(i,":",j)
  

s1 = Smartphone()
print("=================================")
s1.addFeature('Display','6.1 inch')
print("=================================")
s1.setName('Samsung Note 20')
s1.addFeature('Display','6.1 inch')
s1.printDetail()
print("=================================")
s2 = Smartphone('Iphone 12 Pro')
s2.addFeature('Display','6.2 inch')
s2.addFeature('Ram','6 GB')
print('=================================')
s2.printDetail()
s2.addFeature('Display','Amoled panel')
s2.addFeature('Ram','DDR5')
print("=================================")
s2.printDetail()
print("=================================")
```

```python
class Processor:
    def __init__(self, model, thread, core):
        self.model = model
        self.core = core
        self.thread = thread
    def getInfo(self):
        return 'Model:'+self.model+ '\nCores:'+str(self.core)+ '\nThreads:'+ str(self.thread) 
class Intel(Processor):
  def __init__(self,m,c,t,p):
    super().__init__(m,c,t)
    self.p=p
  def getInfo(self):
    print('Model:'+self.model+ '\n'+'Cores:'+str(self.thread)+ '\n'+'Threads:'+ str(self.core)+'\n'+f'Price:{self.p}')
class AMD(Processor):
  def __init__(self,m,t,c,p):
    super().__init__(m,t,c)
  
    self.p=p
  def getInfo(self):
    var=super().getInfo()
    print('Model:'+self.model+ '\n'+'Cores:'+str(self.core)+ '\n'+'Threads:'+ str(self.thread)+'\n'+f'Price:{self.p}')

      
p1 = Intel("Intel i5 10th Gen",6,12,17000)
p2 = AMD("Ryzen 5 3500X",6,6,13800)
p3 = AMD("Ryzen 5 3600",6,12,16900)
print('======================')
p1.getInfo()
print('======================')
p2.getInfo()
print('======================')
p3.getInfo()

```

```python
class Fruit:
    Total_order=0
    
    def __init__(self, Order_ID, weight):
        self.Order_ID=Order_ID
        self.weight=weight
        Fruit.Total_order=Fruit.Total_order+1
    
    def __str__(self):
        return self.Order_ID+", Weight: "+str(self.weight)

class Mango(Fruit):
    def __init__(self,id,wt,name,ms):
       super().__init__(id,wt)
       self.name=name
       self.amount=ms
    def __str__(self):
      var=super().__str__()
      self.t_P=self.weight*self.amount
      return var+','+'Variety: '+ self.name+'\n'+'Total price: '+str(self.t_P)
    def __add__(self,other):
      return 'Total price of the orders are: '+str(self.t_P+other.t_P)
class JackFruit(Fruit):
    def __init__(self,id,wt,ms):
       super().__init__(id,wt)
       self.amount=ms
    def __str__(self):
      var=super().__str__()
      self.t_P=self.weight*self.amount
      return var+','+'\n'+'Total price: '+str(self.t_P)
    def __add__(self,other):
      return 'Total price of the orders are: '+str(self.t_P+other.t_P)


    
m1=Mango("Order Id 1", 5,"GopalVog",250)
print(m1)
m2=Mango("Order Id 2", 5,"HariVanga", 230)
print(m2) 
j1=JackFruit("Order Id 3", 5,250)
print(j1)
j2=JackFruit("Order Id 4", 4,210)
print(j2)
print("Total number of Orders: "+str(Fruit.Total_order))
print("==================")
print(m1+m2)
print("==================")
print(j1+j2)

```
